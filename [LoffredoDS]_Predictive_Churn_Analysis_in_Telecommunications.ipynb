# What is Churn, and why is it important?

Churn is a metric that indicates a company's customer evasion rate. In businesses that provide subscription-based services like Spotify, Netflix, and Amazon Prime, this rate represents the number of canceled subscriptions.
</br>
</br>

Knowing the churn rate is essential because the goal is to minimize it. By understanding this rate over time, companies can validate marketing strategies, pricing, etc. – all to maximize profits and avoid losses.

<br/>
<p align=center>
<img src="https://blogger.googleusercontent.com/img/b/R29vZ2xl/AVvXsEjfvI5kluAw85X6KxJRtoRp1npQ6oWPhr13HRwXZIFpuUnD7S4mG3ahkiHtXsC-A8jjIfvqrpLpE7O8fvbcf4t5WuO7Urm2uYjEImhy7cD3l4Ci23dzgCps4bVJ7NujluetbmckJ0IjXIx_PpkGWn1jaQrZ-xyMQbYizaIY0LR-IXOeGWg1A1F3HhBkrWg/w640-h640/CAPA_EN-01.png" width="50%"></p>
<br/>



Its calculation is given by dividing the number of lost customers by the total number of active customers.
</br>
</br>
$$
Churn\space Rate = \frac{number\space of\space lost\space customers}{total\space number\space of\space active\space customers}
$$
</br>
</br>

This means that if a company had 1000 customers and lost 10, its churn rate would be 1%.
</br>
</br>

Churn Prediction, or the forecast of churn, is extremely important and can be done using machine learning techniques and past customer turnover data. With this, it will be possible, in conjunction with the Marketing department, to take actions to prevent the loss of those customers identified as potential churns. This customer retention can be very beneficial for the company, as the cost of acquiring new customers, given by the Customer Acquisition Cost (CAC) metric, can be up to 15 times higher than the cost spent on campaigns to retain a customer.

## 3. Handle categorical variables

Next, let's handle the categorical variables. For variables with 2 types of entries, such as yes/no, we will use Label Encoding. This technique converts the data into numeric values, and since we have 2 entries, they will become boolean values as they will be represented by 0 or 1. As for categorical attributes that don't fit the above processing – in other words, those with 3 or more distinct entries – we will process them using dummy variables.

```python
# identify variables with 2 labels
binary_var = df_clean.nunique()[df_clean.nunique() == 2].keys().tolist()

# distinguish between categorical and numerical attributes
## identify numeric attributes
num_var = [col for col in df_clean.select_dtypes(['int', 'float']).columns.tolist() if col not in binary_var]

## identify the categorical attributes being those that are not numerical, nor are they the categorical ones already treated in the previous step
cat_var = [col for col in df_clean.columns.tolist() if col not in binary_var + num_var]

# create a copy of the data frame
df_proc = df_clean.copy()

# apply Label Encoding
le = LabelEncoder()
for i in binary_var:
    df_proc[i] = le.fit_transform(df_proc[i])

# apply dummy variables
df_proc = pd.get_dummies(df_proc, columns=cat_var)

# check data frame changes
df_proc.head()
```

```python
# check data frame size
df_proc.shape
```

## 4. Separate the target class from the independent classes

Let's separate the data from the target class, i.e., the `Churn` variable that we aim to predict. ## The `val_model` Function

To start this step, first, I will build a function named `val_model` that constructs a pipeline workflow. This pipeline will standardize the data using the Standard Scaler, classify the data (according to what is provided to the function through the `clf` parameter), and evaluate the model through cross-validation, considering the obtained Recall value. The final result of this function will be to display the average Recall value found in the cross-validation process.

```python
def val_model(X, y, clf, quite=False):
    """
    Performs cross-validation with the training data for a given model.

    # Arguments
      X: Data Frame, contains the independent variables.
      y: Series, vector containing the target variable.
      clf: scikit-learn classifier model.
      quiet: bool, indicating whether the function should print the results or not.

    # Returns
      float, the mean of cross-validation scores.
    """

    # convert variables into arrays
    X = np.array(X)
    y = np.array(y)

    # create pipeline
    ## 1. standardize the data using the StandardScaler.
    ## 2. classify the data.
    pipeline = make_pipeline(StandardScaler(), clf)

    # model evaluation by cross validation
    ## according to Recall value
    scores = cross_val_score(pipeline, X, y, scoring='recall')

    # show mean Recall value and model standard deviation
    if quite == False:
        print("Recall: {:.2f} (+/- {:.2f})".format(scores.mean(), scores.std()))

    # return the average of Recall values obtained in cross-validation
    return scores.mean()
```

## Performance Assessment Metrics

To evaluate the performance of the models, the metric that will be used is the Recall value. This choice stems from the fact that, in the case of imbalanced datasets like the one being studied, even after being treated and balanced, accuracy is not a suitable evaluation metric. This is because achieving a high-accuracy result might lead to a situation where the model has a very low detection rate for churn cases, which does not align with our objective.
<br/>
<br/>

Recall is the metric that provides the best measure for the specific problem under study. This is because, in the case of churn prediction, False Negatives are more detrimental to a company than False Positives. In other words, it's better for the model to err by classifying a customer as churn when they are not, rather than failing to identify a churn customer, which would result in business losses.

Given this context, the goal is to achieve a high Recall rate.

Recall looks at all the churn cases and aims to answer: How much does the model get right? The result is a value between 0 and 1, where closer to 1 is better, as it indicates a low rate of False Negatives.
<br/>
<br/>

Its calculation is given by:
<br/>
<br/>

$${\text{Recall: }}\frac{\text{True Positive}} {\text{True Positive + False Negative)}}$$
<br/>
<br/>

Considering the purpose of this study, another classification model evaluation metric that can be used is the AUC - Area Under the Curve. This metric derives from the ROC - Receiver Operating Characteristic - curve and tells us how well the model can distinguish between two things. In our case, it's about distinguishing between churn and non-churn customers. The AUC value ranges between 0 and 1, where closer to 1 indicates a better model.
<br/>
<br/>

# XGBoost: Hyperparameter Optimization

Due to the nature of the problem we are attempting to solve (as stated in the [General Objective](#scrollTo=5rSx8OkFXz67)), as well as the performance of algorithms in the previous section, XGBoost has a significant potential for improvement through adjustments in its hyperparameters. Given the substantial number of hyperparameters that need to be adjusted, it's necessary to find the optimal value for each of them through separate processes. To begin, a learning rate of 0.1 has been defined. The recommended range is between 0.05 and 0.3, depending on the problem at hand.

Now, we will proceed with the search for the best parameter value for `n_estimators`, which corresponds to the number of decision trees in the algorithm.

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
xgb = XGBClassifier(learning_rate=0.1)

# define dictionary to discover the ideal amount of trees
## in a range from 0 to 500 with increments of 50
param_grid = {'n_estimators':range(0,500,50)}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'n_estimators'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# print best parameter found to 'n_estimators'
print("Best result: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

The best result was achieved with 50 decision trees. Given that the defined increment was 50, we can now narrow down the search around this value with smaller steps.

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
xgb = XGBClassifier(learning_rate=0.1)

# definir dicionário para descobrir a quantidade ideal de árvores
## em um range de 0 a 100 com incremento de 10
param_grid = {'n_estimators':range(0,100,10)}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'n_estimators'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# print best parameter found to 'n_estimators'
print("Best result: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

With this new search, the best value found for the number of trees was 70. With just this adjustment, we have further improved the performance of the XGBoost algorithm. In other words, with only one tuning step, XGBoost has moved up to the 3rd position in the list of models generated in the previous section. We have four more tunings ahead, so it looks promising!

Having found the best parameter value for `n_estimators`, we can now set it in the classifier assignment and proceed to determine the best values for `max_depth`, which defines the depth of the decision trees, and for `min_child_weight`, which is the minimum node weight required to create a new node.

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
## n_estimators: 70
xgb = XGBClassifier(learning_rate=0.1, n_estimators=70)

# set dictionary to find out:
## the ideal depth in a range of 1 to 7 in increments of 1
## the minimum weight of the node in a range from 1 to 4 in increments of 1
param_grid = {'max_depth':range(1,8,1),
              'min_child_weight':range(1,5,1)}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'max_depth' e 'min_child_weight'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# print best parameters found to 'max_depth' e 'min_child_weight'
print("Best results: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

The values obtained for both max_depth and min_child_weight are already set to their minimum possible values, so there's no room for trying to achieve an even better result.

It's worth noting that, up to this point, we have already surpassed the best Recall value of 0.8017 obtained previously with the Logistic Regression model.

We have two more parameters to adjust!

We can incorporate the newly found parameter values and move on to finding the next one: `gamma`. This parameter defines the complexity of the trees in the model.

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
## n_estimators: 70
## max_depth: 1
## min_child_weight: 1
xgb = XGBClassifier(learning_rate=0.1, n_estimators=70, max_depth=1, min_child_weight=1)

# set dictionary to find out:
## the ideal depth in a range of 0.0 to 0.4
param_grid = {'gamma':[i/10.0 for i in range(0,5)]}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'gamma'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# print best parameter found to 'gamma'
print("Best result: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Once again, the best value found is the minimum value. Therefore, we continue by setting this value in the classifier model assignment and proceed to find the optimal value for the learning rate (`learning_rate`).

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
## n_estimators: 70
## max_depth: 1
## min_child_weight: 1
## gamma = 0.0
xgb = XGBClassifier(n_estimators=70, max_depth=1, min_child_weight=1, gamma=0.0)

# set dictionary to discover optimal learning rate
param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.2]}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'learning_rate'
grid_result = grid_search.fit(X_train_rus, y_train_rus)

# print best parameter found to 'learning_rate'
print("Best result: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

# XGBoost: Feature Engineering + Hyperparameter Optimization

As we saw significant improvement not only in the model comparison but also with XGBoost, let's fine-tune it again to see if we can achieve even better performance. We will continue with the same hyperparameter search as before to obtain the best model based on these results.

Let's start by searching for the best value for `n_estimators`, which is the number of decision trees in the algorithm.

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
xgb = XGBClassifier(learning_rate=0.1)

# define dictionary to discover the ideal amount of trees
## in a range from 0 to 300 with increments of 50
param_grid = {'n_estimators':range(0,300,50)}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'n_estimators'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# print best parameter found to 'n_estimators'
print("Best result: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

The best result was achieved with 50 decision trees. Since we used an increment of 50, we can now narrow down the search around this value with smaller steps:

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
xgb = XGBClassifier(learning_rate=0.1)

# define dictionary to discover the ideal amount of trees
## in a range from 0 to 50 with increments of 2
param_grid = {'n_estimators':range(0,50,2)}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'n_estimators'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# print best parameter found to 'n_estimators'
print("Best result: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

With this new search, the best value found for the number of trees was 30. This time, with just this adjustment, we improved the algorithm's performance from 0.7528 to 0.8060. This result would already secure the second place among the baseline models!

We still have four more tuning steps to go, so let's proceed.

Having found the best parameter value for `n_estimators`, we can set it in the classifier assignment and move on to determining the optimal values for `max_depth`, which defines the depth of the decision trees, and for `min_child_weight`, which is the minimum node weight required to create a new node.

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
## n_estimators: 30
xgb = XGBClassifier(learning_rate=0.1, n_estimators=30)

# set dictionary to find out:
## the ideal depth in a range of 1 to 7 in increments of 1
## the minimum weight of the node in a range from 1 to 4 in increments of 1
param_grid = {'max_depth':range(1,8,1),
              'min_child_weight':range(1,5,1)}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'max_depth' e 'min_child_weight'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# print best parameters found to 'max_depth' e 'min_child_weight'
print("Best results: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

We can insert the newly found parameter values and move on to finding the next one: `gamma`. This parameter defines the complexity of the trees in the model.

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
## n_estimators: 30
## max_depth: 1
## min_child_weight: 1
xgb = XGBClassifier(learning_rate=0.1, n_estimators=30, max_depth=1, min_child_weight=1)

# set dictionary to find out:
## the ideal depth in a range of 0.0 to 0.4
param_grid = {'gamma':[i/10.0 for i in range(0,5)]}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'gamma'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# print best parameter found to 'gamma'
print("Best result: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Once again, the best value found is the minimum value. Therefore, we continue by setting this value in the classifier model assignment and proceed to find the optimal value for the learning rate (`learning_rate`).

```python
# instantiate the XGBoost model
## set the learning rate to 0.1
## n_estimators: 30
## max_depth: 1
## min_child_weight: 1
## gamma = 0.0
xgb = XGBClassifier(n_estimators=30, max_depth=1, min_child_weight=1, gamma=0.0)

# set dictionary to discover optimal learning rate
param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.2]}

# set up cross validation with 10 stratified folds
## shuffle=True to shuffle the data before splitting
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuring the search for cross matches with the XGBoost classifier
## tree parameters of param_grid
## scoring: evaluation method by Recall
## n_jobs=1 for parallel fetching (using all available cores)
## cv: validation strategy
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# perform best parameter search for 'learning_rate'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# print best parameter found to 'learning_rate'
print("Best result: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Among the tested values, the one that yielded the best result was a learning rate of 0.01. We could explore even lower values, but as the learning rate decreases, the risk of overfitting the model increases. Therefore, I will stick with this value.

I will now run the test dataset with the created model.

```python
# instantiate the final XGBoost model with the best values found
xgb = XGBClassifier(learning_rate=0.01 , n_estimators=30, max_depth=1, min_child_weight=1, gamma=0.0)

# train the model with training data
xgb.fit(X_fe_train_rus, y_fe_train_rus)

# standardize test data
X_fe_test = scaler.transform(X_fe_test)

# make predictions with test data
y_fe_pred = xgb.predict(X_fe_test)
```
